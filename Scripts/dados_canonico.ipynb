```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
alpha = [1.0, 1.4, 1.8, 2.0]
N = 4096
```

## Para simulação com Campo Externo clrf.py

```python
data = []
for i in range(len(alpha)):
    data.append(pd.read_csv('clrf{}-alpha{}-1000.csv'.format(N, alpha[i])))
```

## Para simulação sem Campo clr.py

```python
#Scripts/clr2048g-alpha1.0-2000-2000.csv
data = []
for i in range(len(alpha)):
    data.append(pd.read_csv('clr{}-alpha{}-4000-4000.csv'.format(N, alpha[i])))
```

```python
T = []; E = []; C = []; M = []; X = []

for i in range (len(data)):
    T.append(data[i]['temperature'])
    E.append(data[i]['energy'])
    C.append(data[i]['specific_heat'])
    M.append(data[i]['magnetization'])
    X.append(data[i]['susceptibility'])
```

```python

f = plt.figure(figsize=(20, 12)); # plot the calculated values
# plt.title(f'1-D RFLRIM, with N={N}, eqSteps={eqSteps}, mcSteps={mcSteps}, $\\Delta$={delta}', fontsize=20);

sp =  f.add_subplot(2, 2, 1 );
for i in range(len(alpha)):
    plt.plot(T[i], E[i],'-o', label=f'$\\alpha$={alpha[i]}');
    plt.xlabel("T", fontsize=20);
    plt.ylabel("U", fontsize=20);
    plt.legend(loc='best', fontsize=15);

sp =  f.add_subplot(2, 2, 2 );
for i in range(len(alpha)):
    plt.plot(T[i], abs(M[i]),'-o', label=f'$\\alpha$={alpha[i]}')
    plt.xlabel("T", fontsize=20);
    plt.ylabel("M ", fontsize=20);
    plt.legend(loc='best', fontsize=15);

sp =  f.add_subplot(2, 2, 3 );
for i in range(len(alpha)):
    plt.plot(T[i], C[i], '-o', label=f'$\\alpha$={alpha[i]}');
    plt.xlabel("T", fontsize=20);
    plt.ylabel("$C$ ", fontsize=20);
    plt.legend(loc='best', fontsize=15);

sp =  f.add_subplot(2, 2, 4 );
for i in range(len(alpha)):
    plt.plot(T[i], X[i], '-o', label=f'$\\alpha$={alpha[i]}');
    plt.xlabel("T", fontsize=20);
    plt.ylabel("$\\chi$", fontsize=20);
    plt.legend(loc='best', fontsize=15);

# plt.savefig(f'Imagens/1d_rfnnim_N{N}_mcS{mcSteps}_eq{eqSteps}.png', bbox_inches='tight')
    
```
